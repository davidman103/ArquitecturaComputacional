```python
import matplotlib.pyplot as plt
import math as m
import numpy as np

#almacen

hilos = np.arange(1,9)
time = [1.989,4.731,5.037,5.879,6.606,6.931,7.112,7.701]
O_N = np.zeros(8)

for i in range(len(hilos)):

  O_N[i] = time[0]/time[i]

plt.plot(hilos,O_N)
plt.xlabel('hilos')
plt.ylabel('t-h1/t-hilos')
plt.title('relacion de tiempos entre hilos')
plt.show
```
